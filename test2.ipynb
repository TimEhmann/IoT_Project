```python
# read dataframe from data/transformer_multivariate_dataframe_v1.csv
df = pd.read_csv('data/transformer_multivariate_dataframe_v1.csv')
df
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
# Get the columns to scale
columns_to_scale = [col for col in df.columns if col != 'date_time_rounded']

# Fit on training data and transform both training and test data
df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])
y_feature_scaler_index = columns_to_scale.index('CO2')
```

```python
# print the values of the scaler
scaler.mean_, scaler.var_
```

```python
import numpy as np

# Example setup
np.random.seed(0)
np_array = np.random.rand(10, 1)  # Your new CO2 values in scaled form


# Create a full-width array with zeros
full_array = np.zeros((10, 50))  # 50 should be replaced with the actual number of columns your scaler was trained on

# Place your CO2 values into the correct column in the full array
y_feature_scaler_index = columns_to_scale.index('CO2')  # You already have this index
full_array[:, y_feature_scaler_index] = np_array.flatten()  # Insert CO2 values into the correct column

# Inverse transform the full array
inverse_transformed = scaler.inverse_transform(full_array)

# Extract only the CO2 column from the inversely transformed array
co2_unscaled = inverse_transformed[:, y_feature_scaler_index]

# co2_unscaled now contains the CO2 values transformed back to their original scaling
print("Original Unscaled CO2 Values:")
print(co2_unscaled)
```
