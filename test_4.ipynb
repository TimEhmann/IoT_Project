```python
import pickle
import torch

device = utils.get_device()

model = utils.load_transformer_model(y_feature="CO2", model_name = 'transformer_multivariate_quarter_hour_26f', device=device)
scaler = utils.load_scaler(y_feature="CO2", model_name = 'transformer_multivariate_quarter_hour_26f')

# Load the last batch from the file
with open('data/last_batch.pkl', 'rb') as f:
    x_batch, device_ids_batch, y_batch = pickle.load(f)

# Convert numpy arrays back to PyTorch tensors
x_batch = torch.tensor(x_batch).to(device)
device_ids_batch = torch.tensor(device_ids_batch).to(device)
y_batch = torch.tensor(y_batch).to(device)

print(x_batch.shape)
input_dim = x_batch.shape[2]

# Evaluate with model
outputs = model(x_batch, device_ids_batch)
actual_batch = y_batch.cpu().numpy().reshape(-1, 1)
predicted_batch = outputs.cpu().numpy().reshape(-1, 1)
zeroes_for_scaler = np.zeros((actual_batch.shape[0], input_dim))

zeroes_for_scaler[:, 2] = actual_batch.flatten()  # Insert actual values into the correct column
inverse_transformed = scaler.inverse_transform(zeroes_for_scaler)
actual_batch_unscaled = inverse_transformed[:, 2]

zeroes_for_scaler[:, 2] = predicted_batch.flatten()  # Insert predicted values into the correct column
inverse_transformed = scaler.inverse_transform(zeroes_for_scaler)
predicted_batch_unscaled = inverse_transformed[:, 2]

predictions = np.array(predicted_batch_unscaled)
actual = np.array(actual_batch_unscaled)

# print dataframe of actual vs predicted with inverse transform
# print multiple accuracy messurements
rmse = np.sqrt(np.mean((actual - predictions)**2))
print(f"Score (RMSE): {rmse:.4f}")
mae = np.mean(np.abs(actual - predictions))
print(f"Score (MAE): {mae:.4f}")
me = np.mean(actual - predictions)
print(f"Score (ME): {me:.4f}")
mape = np.mean(np.abs((actual - predictions) / actual)) * 100
print(f"Score (MAPE): {mape:.4f}%")

pd.DataFrame({'actual': actual, 'predicted': predictions})
```
