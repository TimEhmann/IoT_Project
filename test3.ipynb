```python
input_dim = 22
d_model = 256
nhead = 8
num_layers = 5
dropout_pe = 0.15
dropout_encoder = 0.15
num_devices = 26
device_embedding_dim = 4

for file in os.listdir('models/'):
    if file.endswith('model_v1.pth'):
        if 'hum' in file:
            # hum we are overfitting with the same parameters. Decreasing num_layers by 2
            num_layers -= 2
        elif 'tmp' in file:
            # tmp we are overfitting a little bit with the same parameters. Increasing dropout probability by 0.05 and decreasing num_layers by 1
            num_layers -= 1
            dropout_pe += 0.05
            dropout_encoder += 0.05
        model = TransformerModel(input_dim=input_dim, d_model=d_model, nhead=nhead, num_layers=num_layers, dropout_pe=dropout_pe, dropout_encoder=dropout_encoder, num_devices=num_devices, device_embedding_dim=device_embedding_dim).to(device)
        model.load_state_dict(torch.load('models/' + file, map_location=device))
        save_compressed_model(model, 'models/' + file + '.gz')
        os.remove('models/' + file)

        if 'hum' in file:
            # hum we are overfitting with the same parameters. Decreasing num_layers by 2
            num_layers += 2
        elif 'tmp' in file:
            # tmp we are overfitting a little bit with the same parameters. Increasing dropout probability by 0.05 and decreasing num_layers by 1
            num_layers += 1
            dropout_pe -= 0.05
            dropout_encoder -= 0.05
```

```python
def load_compressed_entire_model(model_path, device):
    temp_path = model_path + '.tmp'
    # Decompress the file to a temporary path
    with gzip.open(model_path, 'rb') as f_in:
        with open(temp_path, 'wb') as f_out:
            shutil.copyfileobj(f_in, f_out)
    
    # Load the entire model
    model = torch.load(temp_path, map_location=device)
    
    # Remove the temporary file
    os.remove(temp_path)
    
    return model

# Example usage
model_path = r'C:\Studium\Semester_6\IoT_Project\models\transformer_multivariate_quarter_hour_23f_tmp_model_v1_test.pth.gz'
device = "cuda" if torch.cuda.is_available() else "cpu"
model = load_compressed_entire_model(model_path, device)
```
