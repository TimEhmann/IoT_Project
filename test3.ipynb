```python
df = deepcopy(df2)

aggregation_level = 'quarter_hour'
window_size = 20
y_feature = 'CO2'
import numpy as np
from sklearn.preprocessing import StandardScaler
from copy import deepcopy

print(df.dtypes)
if aggregation_level == "hour":
    freq='60T'
    df['date_time_rounded'] = df['date_time'].dt.round('60T')
elif aggregation_level == "half_hour":
    freq='30T'
    df['date_time_rounded'] = df['date_time'].dt.round('30T')
elif aggregation_level == "quarter_hour":
    freq='15T'
    df['date_time_rounded'] = df['date_time'].dt.round('15T')
else:
    raise ValueError("Invalid aggregation_level. Please choose one of 'hour', 'half_hour', or 'quarter_hour'.")
print(df.dtypes)
# add time dependent information
# df['weekday'] = df['date_time_rounded'].dt.weekday
# df['month'] = df['date_time_rounded'].dt.month
# df['hour_sin'] = np.sin(2 * np.pi * (df['date_time_rounded'].dt.hour * 3600 + df['date_time_rounded'].dt.minute * 60) / 86400.0)
df['weekday_sin'] = np.sin(2 * np.pi * df['date_time_rounded'].dt.weekday / 7)
df['weekday_cos'] = np.cos(2 * np.pi * df['date_time_rounded'].dt.weekday / 7)
df['month_sin'] = np.sin(2 * np.pi * df['date_time_rounded'].dt.month / 12)
df['month_cos'] = np.cos(2 * np.pi * df['date_time_rounded'].dt.month / 12)
df['time_sin'] = np.sin(2 * np.pi * (df['date_time_rounded'].dt.hour * 3600 + df['date_time'].dt.minute * 60) / 86400.0)
df['time_cos'] = np.cos(2 * np.pi * (df['date_time_rounded'].dt.hour * 3600 + df['date_time'].dt.minute * 60) / 86400.0)
df['semester'] = 'WS22/23'
df.loc[df['date_time_rounded'] >= '2023-03-01', 'semester'] = 'SS23'
df.loc[df['date_time_rounded'] >= '2023-09-01', 'semester'] = 'WS23/24'
df = pd.get_dummies(df, columns=['semester'])

df = df.groupby(['device_id', 'date_time_rounded']).mean().reset_index()

print(df.dtypes)
# create 'consecutive_data_point' thats 1 if the previous data point is 1 hour before the current data point and device_id is the same, else 0
time_delta = 3600 if aggregation_level == 'hour' else 1800 if aggregation_level == 'half_hour' else 900
df['consecutive_data_point'] = (df['date_time_rounded'] - df['date_time_rounded'].shift(1)).dt.total_seconds() == time_delta
df['consecutive_data_point'] = df['consecutive_data_point'].astype(int)

# Identify changes and resets (when the value is '0' or there's a change in 'device_id')
df['reset'] = (df['consecutive_data_point'] == 0) | (df['device_id'] != df['device_id'].shift(1))

# Create a group identifier that increments every time a reset occurs
df['group'] = df['reset'].cumsum()

# Calculate cumulative sum of "1"s within each group
df['consecutive_data_points'] = df.groupby(['device_id', 'group'])['consecutive_data_point'].cumsum() - df['consecutive_data_point']
df['group_size'] = df.groupby(['device_id', 'group'])['consecutive_data_point'].transform('count')

df = df[df['group_size'] > window_size]

# You may want to drop the 'reset' and 'group' columns if they are no longer needed
df.drop(['reset', 'consecutive_data_point', 'consecutive_data_points', 'group_size'], axis=1, inplace=True)

threshold_date = df.sort_values('date_time_rounded', ascending=True)['date_time_rounded'].quantile(0.8)
print('training data cutoff: ', threshold_date)

full_preprocessed_df_unscaled = deepcopy(df)

df['device_id'] = df['device_id'].astype('category').cat.codes

df_train = df[df['date_time_rounded'] < threshold_date]
df_test = df[df['date_time_rounded'] >= threshold_date]

# Create the scaler instance
scaler = StandardScaler()

# Get the columns to scale
columns_to_scale = [col for col in df_train.columns if col not in ['date_time_rounded', 'device_id', 'group']]

# Fit on training data and transform both training and test data
df_train[columns_to_scale] = scaler.fit_transform(df_train[columns_to_scale])
df_test[columns_to_scale] = scaler.transform(df_test[columns_to_scale])
y_feature_scaler_index = columns_to_scale.index(y_feature)

# drop unconvertible columns
print(df_train.dtypes)
df_train.drop(['date_time_rounded'], axis=1, inplace=True)
df_test.drop(['date_time_rounded'], axis=1, inplace=True)

print(df_train.dtypes)
```
